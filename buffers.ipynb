```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
freeFloat = pd.read_csv('extraCompanyDataFinal/freeFloatPercfinal.csv')
freeFloat = freeFloat.iloc[::-1]
freeFloat = freeFloat.reset_index(drop=True)

outShares = pd.read_csv('extraCompanyDataFinal/outstandingSharesFinal.csv')
outShares = outShares.iloc[::-1]
outShares = outShares.reset_index(drop=True)
outShares.head(10)
```

```python
for j in range(len(freeFloat.columns)-1):
    for i in range(len(freeFloat.iloc[:,j+1])-1):
        company = freeFloat.iloc[:,j+1].copy()
        company[i] = round(company[i],4)
        if i%4 != 3:
            
            if company[i+1]>0.15 and abs(company[i+1]-company[i])<0.03:
                company[i+1]=company[i]
            elif company[i+1]<0.15 and abs(company[i+1]-company[i])<0.01:
                company[i+1]=company[i]
            freeFloat.iloc[:,j+1] = company
        

freeFloat = freeFloat.iloc[::-1]
freeFloat = freeFloat.reset_index(drop=True)
freeFloat.head(10)
```

```python
for j in range(len(outShares.columns)-1):
    for i in range(len(outShares.iloc[:,j+1])-1):
        company = outShares.iloc[:,j+1].copy()
        if i%4 != 3:
            if company[i]!=0:
                if (abs(company[i+1]-company[i]))/company[i]<0.01:
                    company[i+1]=company[i]
            outShares.iloc[:,j+1] = company
outShares = outShares.iloc[::-1]
outShares = outShares.reset_index(drop=True)
freeFloat.head(10)
```

```python
freeFloat.to_csv('extraCompanyDataFinal/freeFloatBuffered.csv')
outShares.to_csv('extraCompanyDataFinal/outSharesBuffered.csv')
```
